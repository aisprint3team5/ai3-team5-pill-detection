```python
import os
import models.yolo8.config.path as PATH
import random
import shutil
from pathlib import Path
from tqdm import tqdm

def split_dataset(images_dir, labels_dir, output_dir, train_ratio=0.9):
    images = list(Path(images_dir).glob("*.png"))
    random.shuffle(images)

    split_idx = int(len(images) * train_ratio)
    train_images = images[:split_idx]
    val_images = images[split_idx:]

    for phase, image_list in zip(["train", "val"], [train_images, val_images]):
        phase_image_dir = Path(output_dir) / phase / "images"
        phase_label_dir = Path(output_dir) / phase / "labels"

        phase_image_dir.mkdir(parents=True, exist_ok=True)
        phase_label_dir.mkdir(parents=True, exist_ok=True)

        print(f"[DEV - INFO] Copying {phase} set... ({len(image_list)} images)")
        for img_path in tqdm(image_list):
            label_path = Path(labels_dir) / (img_path.stem + ".txt")

            shutil.copy(img_path, phase_image_dir / img_path.name)
            if label_path.exists():
                shutil.copy(label_path, phase_label_dir / label_path.name)
            else:
                print(f"[WARNING] Label not found for image: {img_path.name}")

if __name__ == "__main__":
    split_dataset(
        images_dir=PATH.TRAIN_IMAGE_DIR,
        labels_dir=PATH.TRAIN_LABEL_DIR,
        output_dir=PATH.TEMP_SPLIT_DIR
    )
```

###raw에서 split 만들기

```python
import os
import shutil
from pathlib import Path

# 원본 경로
src_root = Path("/content/ai3-team5-pill-detection/data/raw/train_annotations")
# 대상 경로
dst_dir = Path("/content/ai3-team5-pill-detection/data/raw/train_labels")

# 대상 폴더 없으면 생성
dst_dir.mkdir(parents=True, exist_ok=True)

# 모든 파일 (폴더 제외) 재귀 탐색
all_files = [f for f in src_root.rglob("*") if f.is_file()]

print(f"[INFO] 총 {len(all_files)}개의 파일을 복사합니다...")

for file_path in all_files:
    dst_path = dst_dir / file_path.name

    # 이름 중복 방지 처리
    counter = 1
    while dst_path.exists():
        dst_path = dst_dir / f"{file_path.stem}_{counter}{file_path.suffix}"
        counter += 1

    shutil.copy(file_path, dst_path)
    print(f"[COPIED] {file_path} → {dst_path}")

print("[DONE] 모든 파일 복사가 완료되었습니다.")
```

```python
import os
import models.yolo8.config.path as PATH
import random
import shutil
from pathlib import Path
from tqdm import tqdm

def split_dataset(images_dir, labels_dir, output_dir, train_ratio=0.9):
    images = list(Path(images_dir).glob("*.png"))
    random.shuffle(images)

    split_idx = int(len(images) * train_ratio)
    train_images = images[:split_idx]
    val_images = images[split_idx:]

    for phase, image_list in zip(["train", "val"], [train_images, val_images]):
        phase_image_dir = Path(output_dir) / phase / "images"
        phase_label_dir = Path(output_dir) / phase / "labels"

        phase_image_dir.mkdir(parents=True, exist_ok=True)
        phase_label_dir.mkdir(parents=True, exist_ok=True)

        print(f"[DEV - INFO] Copying {phase} set... ({len(image_list)} images)")
        for img_path in tqdm(image_list):
            label_path = Path(labels_dir) / (img_path.stem + ".txt")

            shutil.copy(img_path, phase_image_dir / img_path.name)
            if label_path.exists():
                shutil.copy(label_path, phase_label_dir / label_path.name)
            else:
                print(f"[WARNING] Label not found for image: {img_path.name}")

if __name__ == "__main__":
    split_dataset(
        images_dir=PATH.TRAIN_IMAGE_DIR,
        labels_dir=PATH.TRAIN_LABEL_DIR,
        output_dir=PATH.TEMP_SPLIT_DIR
    )
```

###### data.yaml에서 경로 고려하기 -확인 또는 바꿔주기

```python
def update_data_yaml_path(yaml_path, new_path_value):
    with open(yaml_path, "r", encoding="utf-8") as f:
        lines = f.readlines()

    updated_lines = []
    old_path_value = None

    for line in lines:
        if line.strip().startswith("path:"):
            old_path_value = line.strip().split(":", 1)[1].strip()
            updated_lines.append(f"path: {new_path_value}\n")
        else:
            updated_lines.append(line)

    with open(yaml_path, "w", encoding="utf-8") as f:
        f.writelines(updated_lines)

    if old_path_value:
        print(f"[INFO] 기존 path: {old_path_value}")
        print(f"[INFO] 새 path: {new_path_value}")
    else:
        print("[WARNING] 'path:' 항목을 찾을 수 없습니다.")

# 사용 예시
yaml_file = "/content/ai3-team5-pill-detection/data.yaml"
new_path = "data/temp/yolo_split"    #같은 폴더라도 이것은 받은데이터가 아니라 raw에서 가져온 데이터로 스플릿한 것
update_data_yaml_path(yaml_file, new_path)
```

```python
import yaml
import os
import shutil

# 반복 횟수
num_runs = 3    ## ←change this

# 기본 값 정의
base_values = {
    "conf_threshold": 0.5,
    "iou_threshold": 0.4,   # 시작값
    "epochs": 50,
    "patience": 20,
    "image_size": 416,
    "batch_size": 8,
    "test_image_dir": "data/raw/test_images",
    "train_data_yaml": "data.yaml"
}

# 증가시킬 항목과 증가량 정의
increments = {
    "iou_threshold": 0.2,  ### ←change this  # 매 반복마다 설정량 씩 증가
}

yaml_path = "/content/ai3-team5-pill-detection/config/yolo8/yolo_8n.yaml"
main_py_path = "models/yolo8/main.py"

# 반복 실행
for i in range(num_runs):
    print(f"\n🔁 {i+1}번째 실행 시작합니다...")

    # 현재 반복에서 적용할 값 만들기
    current_values = base_values.copy()
    for key, increment in increments.items():
        current_values[key] += increment * i

    # YAML 파일 수정
    with open(yaml_path, 'r') as f:
        config = yaml.safe_load(f)
    config.update(current_values)
    with open(yaml_path, 'w') as f:
        yaml.safe_dump(config, f, sort_keys=False)
    print(f"✅ YAML 업데이트 완료: {current_values}")

    # 모델 실행
    exit_code = os.system(f"python {main_py_path} --model_type 1 --model_size n")
    if exit_code != 0:
        print(f"⚠️ 모델 실행 오류 발생. 코드: {exit_code}")
        continue

    # 결과 폴더 복사
    src = '/content/ai3-team5-pill-detection/outputs'
    dst = '/content/drive/MyDrive'
    def copy_outputs(src_dir, dst_dir):
        base_dst_dir = os.path.join(dst_dir, "outputs")
        if os.path.exists(base_dst_dir):
            j = 1
            while os.path.exists(f"{base_dst_dir}_{j}"):
                j += 1
            base_dst_dir = f"{base_dst_dir}_{j}"
        shutil.copytree(src_dir, base_dst_dir)
        print(f"📂 결과 폴더 복사 완료: {base_dst_dir}")
    copy_outputs(src, dst)

    # 원본 폴더 내부 폴더 삭제
    def delete_inner_folders_only_fixed():
        for item in os.listdir(src):
            item_path = os.path.join(src, item)
            if os.path.isdir(item_path):
                for subitem in os.listdir(item_path):
                    subitem_path = os.path.join(item_path, subitem)
                    if os.path.isdir(subitem_path):
                        shutil.rmtree(subitem_path)
                        print(f"🗑️ 삭제된 하위 폴더: {subitem_path}")
    delete_inner_folders_only_fixed()
```

```python
import os
import shutil

# 복사할 폴더 경로
src_base = '/content/drive/MyDrive'
dst_base = '/content/outputStreamline'

# 목적지 폴더가 없으면 생성
os.makedirs(dst_base, exist_ok=True)

# 1~70번 폴더 복사
for i in range(1, 71):
    src_folder = os.path.join(src_base, f'outputs_{i}')
    dst_folder = os.path.join(dst_base, f'outputs_{i}')
    if os.path.exists(src_folder):
        # 이미 목적지에 폴더가 있으면 지우고 다시 복사
        if os.path.exists(dst_folder):
            shutil.rmtree(dst_folder)
        shutil.copytree(src_folder, dst_folder)
        print(f"복사 완료: {src_folder} -> {dst_folder}")
    else:
        print(f"존재하지 않음: {src_folder}")
```

```python
### mydrive로 이동
import os
import shutil
# 구글드라이브 경로 확인
src = '/content/ai3-team5-pill-detection/outputs'
dst = '/content/drive/MyDrive'
def copy_outputs(src_dir, dst_dir):     # 목표 디렉토리 경로 생성
    base_dst_dir = os.path.join(dst_dir, "outputs")
    if os.path.exists(base_dst_dir):
        i = 1
        while os.path.exists(f"{base_dst_dir}_{i}"):     # 이미 outputs 폴더가 존재하는지 확인
            i += 1         # 숫자 추가하여 폴더 이름을 변경
        base_dst_dir = f"{base_dst_dir}_{i}"
    shutil.copytree(src_dir, base_dst_dir)     # outputs 폴더 복사
    print(f"폴더가 복사되었습니다: {base_dst_dir}")
copy_outputs(src, dst)
```
