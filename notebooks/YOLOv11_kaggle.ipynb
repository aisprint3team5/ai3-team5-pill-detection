# df vs df_pre+ft

```python
df = pd.read_csv('/kaggle/working/ai3-team5-pill-detection/outputs/logs/yolo11/0729_exp/results.csv')
df_pre = pd.read_csv('/kaggle/working/ai3-team5-pill-detection/outputs/logs/yolo11/0730_pretrain_exp/results.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# df, df_pre는 이미 불러와져 있다고 가정

def print_best_map50_info(df, name):
    max_map50 = df['metrics/mAP50(B)'].max()
    max_epoch = df.loc[df['metrics/mAP50(B)'].idxmax(), 'epoch']
    print(f"[{name}] 최고 mAP50: {max_map50:.4f} at epoch {max_epoch}")
    row = df.loc[df['epoch'] == max_epoch]
    display_cols = ['train/box_loss', 'val/box_loss', 'val/cls_loss', 'metrics/precision(B)', 'metrics/recall(B)']
    print(row[display_cols])

print_best_map50_info(df, "df Model")
print_best_map50_info(df_pre, "df_pre Model")

# 손실 및 성능 지표 비교 그래프

fig, axs = plt.subplots(2, 2, figsize=(14,10))

# # train/val box_loss 비교
axs[0,0].plot(df['epoch'], df['train/box_loss'], label='df train/box_loss', color='blue')
axs[0,0].plot(df['epoch'], df['val/box_loss'], label='df val/box_loss', color='cyan')
axs[0,0].plot(df_pre['epoch'], df_pre['train/box_loss'], label='df_pre train/box_loss', color='red')
axs[0,0].plot(df_pre['epoch'], df_pre['val/box_loss'], label='df_pre val/box_loss', color='orange')
axs[0,0].set_title('Train and Val Box Loss')
axs[0,0].set_xlabel('Epoch')
axs[0,0].set_ylabel('Loss')
axs[0,0].legend()

# train/val cls_loss 비교
axs[0,1].plot(df['epoch'], df['train/cls_loss'], label='df train/cls_loss', color='blue')
axs[0,1].plot(df['epoch'], df['val/cls_loss'], label='df val/cls_loss', color='cyan')
axs[0,1].plot(df_pre['epoch'], df_pre['train/cls_loss'], label='df_pre train/cls_loss', color='red')
axs[0,1].plot(df_pre['epoch'], df_pre['val/cls_loss'], label='df_pre val/cls_loss', color='orange')
axs[0,1].set_title('Train and Val Classification Loss')
axs[0,1].set_xlabel('Epoch')
axs[0,1].set_ylabel('Loss')
axs[0,1].legend()

# precision vs recall 비교
axs[1,0].plot(df['epoch'], df['metrics/precision(B)'], label='df Precision', color='blue')
axs[1,0].plot(df['epoch'], df['metrics/recall(B)'], label='df Recall', color='cyan')
axs[1,0].plot(df_pre['epoch'], df_pre['metrics/precision(B)'], label='df_pre Precision', color='red')
axs[1,0].plot(df_pre['epoch'], df_pre['metrics/recall(B)'], label='df_pre Recall', color='orange')
axs[1,0].set_title('Precision vs Recall')
axs[1,0].set_xlabel('Epoch')
axs[1,0].set_ylabel('Score')
axs[1,0].legend()

# mAP50 / mAP50-95 비교
axs[1,1].plot(df['epoch'], df['metrics/mAP50(B)'], label='df mAP50', color='blue')
axs[1,1].plot(df['epoch'], df['metrics/mAP50-95(B)'], label='df mAP50-95', color='cyan')
axs[1,1].plot(df_pre['epoch'], df_pre['metrics/mAP50(B)'], label='df_pre mAP50', color='red')
axs[1,1].plot(df_pre['epoch'], df_pre['metrics/mAP50-95(B)'], label='df_pre mAP50-95', color='orange')
axs[1,1].set_title('mAP50 and mAP50-95')
axs[1,1].set_xlabel('Epoch')
axs[1,1].set_ylabel('mAP')
axs[1,1].legend()

plt.tight_layout()
plt.show()
```

```python
# 1. 1~50 epoch 결과
df_exp1 = pd.read_csv('/kaggle/working/ai3-team5-pill-detection/outputs/logs/yolo11/0730_pretrain_exp3/results.csv')

# 2. 51~70 epoch 결과
df_exp2 = pd.read_csv('/kaggle/working/ai3-team5-pill-detection/outputs/logs/yolo11/0730_pretrain_exp32/results.csv')

# df_exp2의 epoch 컬럼도 51부터 시작하도록 수정
df_exp2['epoch'] = df_exp2['epoch'] + df_exp1['epoch'].max()

df_pre2 = pd.concat([df_exp1, df_exp2], ignore_index=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


def print_best_map50_info(dfs, name):
    max_map50 = dfs['metrics/mAP50(B)'].max()
    max_epoch = dfs.loc[dfs['metrics/mAP50(B)'].idxmax(), 'epoch']
    print(f"[{name}] 최고 mAP50: {max_map50:.4f} at epoch {max_epoch}")
    row = dfs.loc[dfs['epoch'] == max_epoch]
    display_cols = ['train/box_loss', 'val/box_loss', 'val/cls_loss', 'metrics/precision(B)', 'metrics/recall(B)']
    print(row[display_cols])

print_best_map50_info(df_pre1, "df Model")
print_best_map50_info(df_pre2, "df_pre Model")

# 손실 및 성능 지표 비교 그래프

fig, axs = plt.subplots(2, 2, figsize=(14,10))

# # train/val box_loss 비교
axs[0,0].plot(df_pre1['epoch'], df_pre1['train/box_loss'], label='df_pre1 train/box_loss', color='blue')
axs[0,0].plot(df_pre1['epoch'], df_pre1['val/box_loss'], label='df_pre1 val/box_loss', color='cyan')
axs[0,0].plot(df_pre2['epoch'], df_pre2['train/box_loss'], label='df_pre2 train/box_loss', color='red')
axs[0,0].plot(df_pre2['epoch'], df_pre2['val/box_loss'], label='df_pre2 val/box_loss', color='orange')
axs[0,0].set_title('Train and Val Box Loss')
axs[0,0].set_xlabel('Epoch')
axs[0,0].set_ylabel('Loss')
axs[0,0].legend()

# train/val cls_loss 비교
axs[0,1].plot(df_pre1['epoch'], df_pre1['train/cls_loss'], label='df_pre1 train/cls_loss', color='blue')
axs[0,1].plot(df_pre1['epoch'], df_pre1['val/cls_loss'], label='df_pre1 val/cls_loss', color='cyan')
axs[0,1].plot(df_pre2['epoch'], df_pre2['train/cls_loss'], label='df_pre2 train/cls_loss', color='red')
axs[0,1].plot(df_pre2['epoch'], df_pre2['val/cls_loss'], label='df_pre2 val/cls_loss', color='orange')
axs[0,1].set_title('Train and Val Classification Loss')
axs[0,1].set_xlabel('Epoch')
axs[0,1].set_ylabel('Loss')
axs[0,1].legend()

# precision vs recall 비교
axs[1,0].plot(df_pre1['epoch'], df_pre1['metrics/precision(B)'], label='df_pre1 Precision', color='blue')
axs[1,0].plot(df_pre1['epoch'], df_pre1['metrics/recall(B)'], label='df_pre1 Recall', color='cyan')
axs[1,0].plot(df_pre2['epoch'], df_pre2['metrics/precision(B)'], label='df_pre2 Precision', color='red')
axs[1,0].plot(df_pre2['epoch'], df_pre2['metrics/recall(B)'], label='df_pre2 Recall', color='orange')
axs[1,0].set_title('Precision vs Recall')
axs[1,0].set_xlabel('Epoch')
axs[1,0].set_ylabel('Score')
axs[1,0].legend()

# mAP50 / mAP50-95 비교
axs[1,1].plot(df_pre1['epoch'], df_pre1['metrics/mAP50(B)'], label='df_pre1 mAP50', color='blue')
axs[1,1].plot(df_pre1['epoch'], df_pre1['metrics/mAP50-95(B)'], label='df_pre1 mAP50-95', color='cyan')
axs[1,1].plot(df_pre2['epoch'], df_pre2['metrics/mAP50(B)'], label='df_pre2 mAP50', color='red')
axs[1,1].plot(df_pre2['epoch'], df_pre2['metrics/mAP50-95(B)'], label='df_pre2 mAP50-95', color='orange')
axs[1,1].set_title('mAP50 and mAP50-95')
axs[1,1].set_xlabel('Epoch')
axs[1,1].set_ylabel('mAP')
axs[1,1].legend()

plt.tight_layout()
plt.show()
```
