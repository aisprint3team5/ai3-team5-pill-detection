### YOLO v8 모델 구동

#### main.py 내용바꾸기

```python
# models/yolo8/main.py

from enums.yolo8_weight import Yolo8Weight
from config.config import load_config
import config.path as PATH
from detector.YOLOv8_detector import YOLOV8Detector
from pipeline.inference import run_inference
from utils.visualize import visualize_detection
from utils.to_submission_format import to_submission_format
import os
import sys

if __name__ == "__main__":
    # key=value 형태로 인자 파싱
    cli_args = dict(arg.split("=") for arg in sys.argv[1:])

    model_type = int(cli_args["model_type"])
    model_size = cli_args["model_size"]
    use_p6 = cli_args.get("use_p6", "False").lower() == "true"

    model_enum = {
        1: Yolo8Weight.DETECT,
        2: Yolo8Weight.SEGMENT,
        3: Yolo8Weight.POSE,
        4: Yolo8Weight.CLASSIFY,
        5: Yolo8Weight.ORIENT
    }[model_type]

    model_file_name = model_enum.model_filename(model_size, use_p6)
    config_filename = model_enum.config_filename(model_size)

    print(f"선택된 모델: {model_file_name}")
    print(f"config/yolo8/{config_filename} 구성파일을 로드합니다")

    config_path = PATH.YOLO8_YAML_PATH / config_filename
    if not config_path.exists():
        raise FileNotFoundError(f"구성 파일이 존재하지 않습니다: {config_path}")

    config = load_config(config_path)

    # 구성값 추출
    conf_threshold = config["conf_threshold"]
    train_data_yaml = PATH.ROOT_DIR / config["train_data_yaml"]
    train_epoch = config["epochs"]
    train_patience = config["patience"]
    image_size = config["image_size"]
    batch_size = config["batch_size"]
    test_image_dir = config["test_image_dir"]

    model_path = PATH.MODEL_PATH / model_file_name
    detector = YOLOV8Detector(model_path=model_path, conf_threshold=conf_threshold)

    detector.train(
        data_yaml_path=train_data_yaml,
        epochs=train_epoch,
        patience=train_patience,
        imgsz=image_size,
        batch_size=batch_size
    )

    results = run_inference(detector, test_image_dir)
    print(f"[INFO] Detection 완료. 총 결과 수: {len(results)}")

    for idx, result in enumerate(results):
        save_path = f"{PATH.VISUALIZATION_SAVE_PATH}/result_{idx}.jpg"
        visualize_detection([result], save_path=save_path, show=False)

    test_result, timestamp = detector.test(source_path=test_image_dir)
    to_submission_format(test_result, timestamp)
```

```python
# 파일 내용 통째로 CLI 인자 기반으로 수정
file_path = "/content/ai3-team5-pill-detection/models/yolo8/main.py"

cli_code = '''
from enums.yolo8_weight import Yolo8Weight
from config.config import load_config
import config.path as PATH
from detector.YOLOv8_detector import YOLOV8Detector
from pipeline.inference import run_inference
from utils.visualize import visualize_detection
from utils.to_submission_format import to_submission_format
import os
import sys

if __name__ == "__main__":
    # key=value 형태로 인자 파싱
    cli_args = dict(arg.split("=") for arg in sys.argv[1:])

    model_type = int(cli_args["model_type"])
    model_size = cli_args["model_size"]
    use_p6 = cli_args.get("use_p6", "False").lower() == "true"

    model_enum = {
        1: Yolo8Weight.DETECT,
        2: Yolo8Weight.SEGMENT,
        3: Yolo8Weight.POSE,
        4: Yolo8Weight.CLASSIFY,
        5: Yolo8Weight.ORIENT
    }[model_type]

    model_file_name = model_enum.model_filename(model_size, use_p6)
    config_filename = model_enum.config_filename(model_size)

    print(f"선택된 모델: {model_file_name}")
    print(f"config/yolo8/{config_filename} 구성파일을 로드합니다")

    config_path = PATH.YOLO8_YAML_PATH / config_filename
    if not config_path.exists():
        raise FileNotFoundError(f"구성 파일이 존재하지 않습니다: {config_path}")

    config = load_config(config_path)

    # 구성값 추출
    conf_threshold = config["conf_threshold"]
    train_data_yaml = PATH.ROOT_DIR / config["train_data_yaml"]
    train_epoch = config["epochs"]
    train_patience = config["patience"]
    image_size = config["image_size"]
    batch_size = config["batch_size"]
    test_image_dir = config["test_image_dir"]

    model_path = PATH.MODEL_PATH / model_file_name
    detector = YOLOV8Detector(model_path=model_path, conf_threshold=conf_threshold)

    detector.train(
        data_yaml_path=train_data_yaml,
        epochs=train_epoch,
        patience=train_patience,
        imgsz=image_size,
        batch_size=batch_size
    )

    results = run_inference(detector, test_image_dir)
    print(f"[INFO] Detection 완료. 총 결과 수: {len(results)}")

    for idx, result in enumerate(results):
        save_path = f"{PATH.VISUALIZATION_SAVE_PATH}/result_{idx}.jpg"
        visualize_detection([result], save_path=save_path, show=False)

    test_result, timestamp = detector.test(source_path=test_image_dir)
    to_submission_format(test_result, timestamp)
'''

with open(file_path, "w") as f:
    f.write(cli_code.strip())
```

####:드뎌!꺆! 모델 v8 돌리기 x3 & 결과 저장/ 이후 원본 하위폴더 내용 비우기

```python
import yaml
import os
import shutil

# 반복 횟수
num_runs = 3    ## ←change this

# 기본 값 정의
base_values = {
    "conf_threshold": 0.5,
    "iou_threshold": 0.4,   # 시작값
    "epochs": 50,
    "patience": 20,
    "image_size": 416,
    "batch_size": 8,
    "test_image_dir": "data/raw/test_images",
    "train_data_yaml": "data.yaml"
}

# 증가시킬 항목과 증가량 정의
increments = {
    "iou_threshold": 0.2,  ### ←change this  # 매 반복마다 설정량 씩 증가
}

yaml_path = "/content/ai3-team5-pill-detection/config/yolo8/yolo_8n.yaml"
main_py_path = "models/yolo8/main.py"

# 반복 실행
for i in range(num_runs):
    print(f"\n🔁 {i+1}번째 실행 시작합니다...")

    # 현재 반복에서 적용할 값 만들기
    current_values = base_values.copy()
    for key, increment in increments.items():
        current_values[key] += increment * i

    # YAML 파일 수정
    with open(yaml_path, 'r') as f:
        config = yaml.safe_load(f)
    config.update(current_values)
    with open(yaml_path, 'w') as f:
        yaml.safe_dump(config, f, sort_keys=False)
    print(f"✅ YAML 업데이트 완료: {current_values}")

    # 모델 실행
    exit_code = os.system(f"python {main_py_path} --model_type 1 --model_size n")
    if exit_code != 0:
        print(f"⚠️ 모델 실행 오류 발생. 코드: {exit_code}")
        continue

    # 결과 폴더 복사
    src = '/content/ai3-team5-pill-detection/outputs'
    dst = '/content/drive/MyDrive'
    def copy_outputs(src_dir, dst_dir):
        base_dst_dir = os.path.join(dst_dir, "outputs")
        if os.path.exists(base_dst_dir):
            j = 1
            while os.path.exists(f"{base_dst_dir}_{j}"):
                j += 1
            base_dst_dir = f"{base_dst_dir}_{j}"
        shutil.copytree(src_dir, base_dst_dir)
        print(f"📂 결과 폴더 복사 완료: {base_dst_dir}")
    copy_outputs(src, dst)

    # 원본 폴더 내부 폴더 삭제
    def delete_inner_folders_only_fixed():
        for item in os.listdir(src):
            item_path = os.path.join(src, item)
            if os.path.isdir(item_path):
                for subitem in os.listdir(item_path):
                    subitem_path = os.path.join(item_path, subitem)
                    if os.path.isdir(subitem_path):
                        shutil.rmtree(subitem_path)
                        print(f"🗑️ 삭제된 하위 폴더: {subitem_path}")
    delete_inner_folders_only_fixed()
```
